```python
import requests
import json

def query_earthquake_api(method, params):
    base_url = "https://earthquake.usgs.gov/fdsnws/event/1/"
    url = base_url + method
    response = requests.get(url, params=params)
    return response.json()

def main():
    method = input("Enter the method (e.g., query): ")
    params = {}
    while True:
        key = input("Enter parameter key (press Enter to stop): ").strip()
        if not key:
            break
        value = input(f"Enter value for parameter {key}: ").strip()
        params[key] = value
    result = query_earthquake_api(method, params)
    print(json.dumps(result, indent=4))

if __name__ == "__main__":
    main()
```

```python
import requests
import json

def query_earthquake_api(params):
    base_url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
    response = requests.get(base_url, params=params)
    return response.json()

def main():
    # Example parameters to query earthquake data
    params = {
        "format": "geojson",  # Specify the output format
        "starttime": "2024-03-08",  # Limit to events on or after the specified start time
        "endtime": "2024-03-09",  # Limit to events on or before the specified end time
        "minmagnitude": "0"  # Limit to events with a magnitude larger than the specified minimum
    }

    result = query_earthquake_api(params)
    print(json.dumps(result, indent=4))

if __name__ == "__main__":
    main()
```
